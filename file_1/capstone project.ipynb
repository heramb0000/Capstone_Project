## lets scrape for Las Vegas

```python
 driver = webdriver.Chrome()
driver.maximize_window()
```

```python
driver.get("https://homefinder.com/")
time.sleep(2)  # Wait for the page to load
```

```python
search_location=driver.find_element(By.XPATH,"/html/body/div/div/div/div/section/div/header/div/div/div/form/div/div[1]/div/div/input") 
search_location.send_keys("Las Vegas, NV")
```

```python
click_tab= driver.find_element(By.XPATH,'/html/body/div/div/div/div/section/div/header/div/div/div/form/div/div[1]/div/div/div[2]/button/span')
click_tab.click()
time.sleep(3)
```

```python
# fetching url to open each jobs
estate_url = []
start=0
end=5
for page in range(start,end):
    url= driver.find_elements(By.XPATH,'//div[@class="results-items mt-2"]/div/a')
    for i in url:
        estate_url.append(i.get_attribute('href')) # to retrieve any link we use get_attribute
    li_index = 3 + page
    next_button_xpath = f'/html/body/div[1]/div/div/section/div[1]/div[4]/div[1]/div[3]/nav/ul/li[{li_index}]/a'
    next_button = driver.find_element(By.XPATH, next_button_xpath)
    next_button.click()   
    time.sleep(4)

        
```

```python
# fetching url to open each jobs
start=0
end=10
for page in range(start,end):
    url= driver.find_elements(By.XPATH,'//div[@class="results-items mt-2"]/div/a')
    for i in url:
        estate_url.append(i.get_attribute('href')) # to retrieve any link we use get_attribute
    
    
    next_button = driver.find_element(By.XPATH,'/html/body/div[1]/div/div/section/div[1]/div[4]/div[1]/div[3]/nav/ul/li[7]/a')
    next_button.click()
    time.sleep(4)
```

```python
Bhk = []
Bed = []
Property_Type = []
Bathroom = []
Address = []
City = []
State_code = []
Country = []
Landmark = []
Year_Built = []
Price_per_sqft  = []
Buy_total_price = []
Carpetarea = []

for i in estate_url:
    driver.get(i)
    time.sleep(3)
    
    try:
        bhk_detail = driver.find_element(By.XPATH,'/html/body/div[1]/div/div/section/div[3]/div/div/div/div/div/div[1]/div[4]/ul/li[1]')
        Bhk.append(bhk_detail.text[0:1])
    except NoSuchElementException:
        Bhk.append(np.nan)
        

    try:
        bed_detail = driver.find_element(By.XPATH,'/html/body/div[1]/div/div/section/div[3]/div/div/div/div/div/div[1]/div[4]/ul/li[1]')
        Bed.append(bed_detail.text[0:1])
    except NoSuchElementException:
        Bed.append(np.nan)

    
    try:
        prop_detail = driver.find_element(By.XPATH,'/html/body/div[1]/div/div/section/div[3]/div/div/div/div/div/div[1]/div[2]/p')
        Property_Type.append(prop_detail.text)      
    except NoSuchElementException:
        Property_Type.append(np.nan)
        
    try:
        bathroom_detail = driver.find_element(By.XPATH,'/html/body/div[1]/div/div/section/div[3]/div/div/div/div/div/div[1]/div[4]/ul/li[2]')
        Bathroom.append(bathroom_detail.text[0:1])
    except NoSuchElementException:
        Bathroom.append(np.nan)
        
    try:
        add_detail = driver.find_element(By.XPATH,'/html/body/div[1]/div/div/section/div[3]/div/div/div/div/div/div[1]/div[3]/div[1]/h1')
        Address.append(add_detail.text)
    except NoSuchElementException:
        Address.append(np.nan)
    
    try:
        city_detail = driver.find_element(By.XPATH,'/html/body/div[1]/div/div/section/div[3]/div/div/div/div/div/div[1]/div[3]/div[1]/h1/div[2]')
        City.append(city_detail.text[0:9])
    except NoSuchElementException:
        City.append(np.nan)

    try:
        state_details = driver.find_element(By.XPATH,'/html/body/div[1]/div/div/section/div[3]/div/div/div/div/div/div[1]/div[3]/div[1]/h1/div[2]')
        State_code.append(state_details.text[-8:-6])
    except NoSuchElementException:
        State_code.append(np.nan)

    
    try:
        landmark = driver.find_element(By.XPATH,'/html/body/div[1]/div/div/section/div[4]/div[1]/section[5]/div/div[1]/div/table/tbody/tr[1]/td[1]')
        Landmark.append(landmark.text)
    except NoSuchElementException:
        Landmark.append(np.nan)

    try:
        year = driver.find_element(By.XPATH,'/html/body/div[1]/div/div/section/div[4]/div[1]/section[3]/div[3]/div/div[9]/div/div[2]')
        Year_Built.append(year.text)       
    except NoSuchElementException:
        Year_Built.append(np.nan)
        
    try:
        psq = driver.find_element(By.XPATH,'/html/body/div[1]/div/div/section/div[4]/div[1]/section[3]/div[3]/div/div[8]/div/div[2]')
        Price_per_sqft.append(psq.text)
    except NoSuchElementException:
        Price_per_sqft.append(np.nan)
        
    try:
        total = driver.find_element(By.XPATH,'/html/body/div[1]/div/div/section/div[3]/div/div/div/div/div/div[2]/div/div[1]/p[2]')
        Buy_total_price.append(total.text[:-24])
    except NoSuchElementException:
        Buy_total_price.append(np.nan)
    try:
        carpetarea = driver.find_element(By.XPATH,'/html/body/div[1]/div/div/section/div[3]/div/div/div/div/div/div[1]/div[4]/ul/li[3]')
        Carpetarea.append(carpetarea.text)   
    except NoSuchElementException:
        Carpetarea.append(np.nan)
```
